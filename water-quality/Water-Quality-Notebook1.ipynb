```python
excel = pd.read_excel("domestic2003.xlsx")
columnnames = excel.columns.to_list()
columnnames
emptydf = pd.DataFrame(columns=columnnames)
emptydf
```

```python
def gatherdata(excellist):
    global emptydf
    for excelfile in excellist:
        file = pd.read_excel(excelfile)
        emptydf = pd.concat([emptydf, file], axis=0)
    return emptydf
```

```python
lists = ["domestic2003.xlsx", "domestic2006.xlsx", "domestic2009.xlsx", "domestic2012.xlsx"]
domestic = gatherdata(lists)
domestic
```
