```python
import wbdata
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
indicators = {'EN.ATM.CO2E.KT': 'Total CO2 Emissions(kt)',
              'EN.ATM.CO2E.KD.GD': 'kg CO2 Per 10 Dollars USD GDP', 'EN.ATM.CO2E.PC': "CO2 Tons Per Capita"}
g7 = ['CAN', 'GBR', 'USA', 'FRA', 'ITA', 'DEU', 'JPN']

df = wbdata.get_dataframe(indicators, country=g7, convert_date=True)

df.reset_index(level=['country', 'date'], inplace=True)

df["CO2 Tons Per Capita"] = df.apply(lambda x: x["CO2 Tons Per Capita"]*1000, axis=1)
df = df.rename(columns={"CO2 Tons Per Capita": "CO2 kg Per Capita", "date": "Date"})
```

```python
df
```

```python
fig1 = px.line(df, x="Date", y="Total CO2 Emissions(kt)", color='country')
fig2 = px.line(df, x="Date", y="kg CO2 Per 10 Dollars USD GDP", color='country')
fig3 = px.line(df, x="Date", y="CO2 kg Per Capita", color='country', title = 'G7 Nations CO2 Emissions Per Capita')
fig1.show()
fig2.show()
fig3.show()
```
