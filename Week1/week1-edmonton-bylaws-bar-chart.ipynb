# Data Visualizations 

## Week 1 -  Bar Charts of Bylaw Infractions

Like most cities, the [City of Edmonton](https://www.edmonton.ca) employs bylaw officers to enforce the rules by issuing tickets. Tickets are issued for infractions such as having ice on your walkway in wintertime, weed infestations, graffiti, or running an unlicensed business. Citizens can even report each other for infractions using the [City of Edmonton's Online Bylaw Complaint Service](https://www.edmonton.ca/city_government/bylaws/online-bylaw-complaint-service.aspx).

Bylaw infractions and investigations can be viewed on a [map](https://data.edmonton.ca/Community-Services/Bylaw-Infractions-Map-View/em7g-s625) or as a [data table](https://data.edmonton.ca/Community-Services/Bylaw-Infractions/xgwu-c37w/data). This notebook will use this dataset to explore Edmonton's bylaw enforcement, and answer questions such as:

* How consistent has the number of complaints been over time?
* How do complaints?

Below is one way you can use Python to create a set of graphs to explore these questions. We'll use the [pandas](https://pandas.pydata.org) library that lets us work with data.

```python
import pandas as pd
df = pd.read_csv('https://data.edmonton.ca/api/views/xgwu-c37w/rows.csv?accessType=DOWNLOAD', low_memory=False)
df
```

We see that there are many rows of data under various columns. Let's look at the number of complaints (the `COUNT` column) over time.

```python
#df.drop(df[df['YEAR'] == 2020].index, inplace=True)
# count bylaw violations by month:
df_year_month_sum = df.groupby(['YEAR', 'MONTH_NUMBER'])['COUNT'].sum().reset_index()
# create well-formatted column of datetimes:
df_year_month_sum['Date'] = pd.to_datetime(df_year_month_sum.YEAR*100 + df_year_month_sum.MONTH_NUMBER, format='%Y%m')
df_year_month_sum
```

### Visualizing

We can now create a visualization of bylaw infractions over time using [Plotly Express](https://plotly.com/python/bar-charts).

```python
import plotly.express as px
fig = px.bar(df_year_month_sum, x='Date', y='COUNT', 
             title='Edmonton Bylaw Complaints Over Time')
fig.show()
```

We can also group by year or by month.

```python
fig = px.bar(df_year_month_sum.groupby('YEAR').sum().reset_index(), x='YEAR', y='COUNT', 
             title='Edmonton Bylaw Complaints by Year')
fig.show()
```

```python
fig = px.bar(df_year_month_sum.groupby('MONTH_NUMBER').sum().reset_index(), x='MONTH_NUMBER', y='COUNT', 
             title='Edmonton Bylaw Complaints by Month')
fig.show()
```

We made these visualizations with only a few lines of code.
