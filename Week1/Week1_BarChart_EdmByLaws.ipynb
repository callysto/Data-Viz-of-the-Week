## Bar Charts, Histogramsand Bylaw Infractions

Like most cities, the city of Edmonton employs many bylaw enforcement officers to enforce the city's rules, or bylaws, by issuing tickets. These tickets are issued for infractions such as: having ice on your walkway in wintertime, weed infestation on your property in the summertime, doing graffiti or running an unlicensed business. Private citizens can even report on each other for such infractions using the [city's website](https://www.edmonton.ca/city_government/bylaws/online-bylaw-complaint-service.aspx).

Now, imagine the city of Edmonton, wanting to review their policies and procedures for bylaw enforcement, came to you asking for an overview of bylaw infractions over the past few years. They want know two things:
* How consistent has the number of complaints been over time?
* What is the total number of each type of complaint receievd last year?

To answer the above you could use this publically available [dataset](https://data.edmonton.ca/Community-Services/Bylaw-Infractions-Map-View/em7g-s625) on the issue. This notebook will use this dataset to explore some features of Edmonton's Bylaw enforcement.

After downloading the dataset there are many ways to proceed. However, to start exploring data, it is always a good idea to create some charts or graphs to give you a glimpse into what information the data has. Pictures won't tell you everything but they can get you started on where to look. Below is one way you can use python to create a set of graphs to explore an answer to the questions the city has.

```python
#Import python modules to help with data

#Pandas module lets us work with tabular data easily
import pandas as pd
#Matplotlip is a standard plotting tool for python 
import matplotlib.pyplot as plt
#Seaborn extends matplotlib - we will use it to change the default plotting parameters and make prettier plots 
import seaborn as sns

""" Microsoft(MS) and Macintosh computers have different ways for writing down file paths.
 MS computers will use backslashes, "\", between directories or folders on a computer (e.g. C:\Program Files\Steam). 
 Mac (and linux) machines will use forward slashes, "/", for the same purpose (eg: /documents/homework/math)."""
# We will use the os module below to write a file path that will work on both Macs and PCs.
import os
```

```python
#Import the city of edmonton data from a csv (comma separated values) file into a pandas dataframe:
path = os.path.join('resources', 'Bylaw_Infractions_-_Map_View.csv')
#Output the path just to make sure it worked:
print(path)
```

```python
"""Pandas dataframes are spreadsheet-like objects we can use to read and manipulate data.
Below we make a dataframe by reading in the csv found at path. Setting 'low_memory=False' suppresses a warning about
columns having mixed data types. We don't need to worry about that today."""
# Create a pandas 'dataframe' from the csv saved at 'path':
df = pd.read_csv(path, low_memory=False)
# display the first 5 entries:
df.head()

#note: NaN means "Not a Number". Often this means there was no value for that entry in the original csv.
```

### Results!
If you see 5 rows of data under various columns that means we have our first real success! Now we could try to manually sort through the data. Sometimes manually looking through can be helpful but it is usually too impractical. Let's isolate which columns we want to answer the questions. The date columns will certainly be useful for the first question. For the second question, the 'COMPLAINT' column would allow us to look at the counts by type. For both questions, we should keep the 'COUNT' column so we don't undercount duplicates.

```python
# Remove the incomplete 2020 data
df.drop(df[df['YEAR'] == 2020].index, inplace=True)

""" The line below sums the number of counts for each year
 groupby(['YEAR', 'MONTH_NUMBER']) groups all the rows that share the same entry in the YEAR and MONTH_NUMBER column
 ['COUNT'].sum() ensures what is returned keeps the COUNT variable and returns the total 
 .reset_index(), in this case, is just to make sure the object returned is still a pandas dataframe"""
# remove unwanted columns and count bylaw violations by month:
df_year_month_sum = df.groupby(['YEAR', 'MONTH_NUMBER'])['COUNT'].sum().reset_index()


""" Dates are notoriously difficult to work with for plotting or data analysis.
 The code below creates a new column 'YEAR_MONTH' that holds a special 'datetime' variable which we will use for our plot
 First the code uses the year and the month to transfer the YYYY and MM data into one int of format YYYYMM.
 Finally it creates a new datetime variable using the date information in the YYYYMM format we created.
 When dealing with dates that aren't in the exact format you want there are a lot of methods. Google is your friend."""
# create well formatted column of datetimes:
df_year_month_sum['YEAR_MONTH'] = pd.to_datetime(df_year_month_sum.YEAR*100 + df_year_month_sum.MONTH_NUMBER, format='%Y%m')

# printing the head of the resulting data frames lets us confirm our work.
print(df_year_month_sum.head())
```

```python
# use default seaborn settings:
sns.set()

# create three rows of plots in one column with size indicated:
fig1, axs = plt.subplots(3, 1, constrained_layout=True, figsize=[8.0,13.0])

""" axs is an array holding the 3 subplots.
axs[0] is the top plot, axs[2] is the third. We apply pyplot histograms to each.
df_year_month_sum.YEAR_MONTH specifies the x axis information, in this case our date time
bins holds the number of bars we want. In our case we go yearly(10), semi-yearly(20), and quarterly(40)
weights is y axis information associated with each entry for the x-axis. 
If weights parameter is not specified the hist() function will count the number of entries in each bin to use as weights"""

# plot the subplots and include titles and labeled axes:
_ = axs[0].hist(df_year_month_sum.YEAR_MONTH, bins=10, weights=df_year_month_sum.COUNT)
axs[0].set_title('Bylaw Complaints - Yearly')
axs[0].set_ylabel('Number of Complaints')
axs[0].set_xlabel('Time Period')

_ = axs[1].hist(df_year_month_sum.YEAR_MONTH, bins=20, weights=df_year_month_sum.COUNT)
axs[1].set_title('Bylaw Complaints - Semi-Yearly')
axs[1].set_ylabel('Number of Complaints')
axs[1].set_xlabel('Time Period')

_ = axs[2].hist(df_year_month_sum.YEAR_MONTH, bins=40, weights=df_year_month_sum.COUNT)
axs[2].set_title('Bylaw Complaints - Quarterly')
axs[2].set_ylabel('Number of Complaints')
axs[2].set_xlabel('Time Period')

fig1.suptitle('Bylaw Complaints in Edmonton (2011-2019)')

# display the plot:
fig1.show()

# save the plot:
plt.savefig('histograms.png')
```

## Histograms

We have now, with only a little bit of code, transformed a dataset to express summarize the information we wanted and created a set of reasonably clear histograms showing how many bylaw complaints are created in varied time periods. These histograms group together complaints by the time they were received and the height represents the number of complaints in the time period indicated. In general, histograms are used to group together similar quantitative categories and use the height to represent the size of that group.

Take care when interpreting histograms. It is clear from the above three diagrams that how we grouped the data or the number of bins we selected really changed how the graph appeared. Be very aware of the bin sizes when you see data presented this way. If someone has an agenda and is trying to create a convincing argument, it is very easy to manipulate the visuals by manipulating the bin sizes to create the appearance of a trend without outright lying. You will find that as you tend to less bins, or larger bin sizes you have, the smoother the histogram tends to be.

These histograms seem to show that the number of complaints is stable year to year. It also shows there is variation when you look at smaller chunks of time. What do you think could cause this? Could a seasonal component affect the number of complaints? There are some categories (such as weeds and snow on walkways) that are tied to seasons. Perhaps they are the cause of the variation.

```python
# create a new dataframe containing only the complaints from 2019:
df_19_counts = df[df['YEAR'] == 2019]

"""The long line of code below does a few things.
First it groups by 'COMPLAINT' similar to how we grouped by month and year above. 
next ['COUNT'].sum() sums the COUNT column that share the same COMPLAINT column
reset_index(), as before ensures that still have a pandas dataframe in the end"""
# count the number of each type of complaint:
df_19_counts = df_19_counts.groupby('COMPLAINT')['COUNT'].sum().reset_index()

# print the result:
print(df_19_counts)
```

```python
# make the figure long so the labels have room:
plt.figure(figsize=(11,5))

# reate a bar chart from our dataframe:
_ = plt.bar(df_19_counts['COMPLAINT'], df_19_counts['COUNT'])
plt.xlabel('Type of Complaint')
plt.ylabel('Number of Complaints')
plt.title('Breakdown of Complaints by Type - 2019')

# save the bar cahrt
plt.savefig('bar_chart.png')

# show our bar chart:
plt.show()
```

## Bar Graphs

A bar graph, like the one above, the columns represent different categories of things. Remember, in the histograms earlier the columns represented an interval of time. In this bar graph the columns represent the different categories of bylaw infractions.
